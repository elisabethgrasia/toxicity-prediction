### Convert SMILES into Morgan Fingerprints

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
```

```python
data = {
    "smiles": ["CCO", "CCN", "CCC", "CCCl"],
    "label": [0, 1, 0, 1]
}

df = pd.DataFrame(data)
```

```python
df["mol"] = df["smiles"].apply(Chem.MolFromSmiles)

# before converting to fingerprints
df = df[df["mol"].notnull()]
```

```python
# create fingerprints generator

morgan_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
```

```python
df["fp"] = df["mol"].apply(lambda mol: morgan_gen.GetFingerprint(mol))
```

```python
# convert fingerprints to NumPy array for ML inputs

X = np.array([np.array(fp) for fp in df["fp"]])
y = df["label"].values
```

```python
X.shape
```

```python
# ML model implementation

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```
